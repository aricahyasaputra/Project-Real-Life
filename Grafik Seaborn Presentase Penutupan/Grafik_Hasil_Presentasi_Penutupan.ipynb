```python
# memanggil modul Pandas and Seaborn  
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import seaborn as sns
from google.colab import files
```

```python
#plt.figure(figsize=(12,8))
nama = ["Line Transek 1", "Line Transek 2", "Line Transek 3"]
jumlah = [18.63, 8.63, 5.36]
color = ["cornflowerblue", "orange", "darkgrey"]
df = pd.DataFrame(list(zip(nama, jumlah)),
               columns =['Name', 'val'])
ax = sns.barplot(x='Name', y='val', data=df, palette=color)
ax.set(xlabel=None, ylabel=None)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.yaxis.set_major_formatter(PercentFormatter(100))
for p in ax.patches:
        ax.annotate('{}'.format(p.get_height()), (p.get_x()+0.40, p.get_height()),
                    ha='center', va='bottom',
                    color= 'black')
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)

change_width(ax, .20)
ax.set(ylim=(0, 20))
plt.legend([],[], frameon=False)

plt.savefig('name.png',bbox_inches='tight')
files.download('name.png')
```
